```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([12,23,34,45,67])
y = np.array([34,56,76,79,89])
z = np.array([87,76,56,48,32])
```

# line plot

```python
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.title('Line Plot')
# markers = o, >, <, ^, D
plt.plot(x, y, marker = 'o', mfc = 'r', color = 'black',
        ms = 7, ls = '--')
```

```python
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.title('Line Plot')
# markers = o, >, <, ^, D
plt.plot(x, z, marker = 'o', mfc = 'black', color = 'green',
        ms = 7, ls = '--')
```

```python
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.title('Line Plot')
# markers = o, >, <, ^, D
plt.plot(x, y, marker = 'o', mfc = 'r', color = 'black',
        ms = 7, ls = '--', label = '2023')
plt.plot(x, z, marker = 'o', mfc = 'black', color = 'green',
        ms = 7, ls = '--',label = '2024')
plt.grid()
plt.legend()
```

```python
# subplot
```

```python

plt.subplot(2,2,2)
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.title('Line Plot')
# markers = o, >, <, ^, D
plt.plot(x, y, marker = 'o', mfc = 'r', color = 'black',
        ms = 7, ls = '--')

plt.subplot(2,2,1)
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.title('Line Plot')
# markers = o, >, <, ^, D
plt.plot(x, z, marker = 'o', mfc = 'black', color = 'green',
        ms = 7, ls = '--')
```

# Bar Plot

```python
courses = ['java','python','R','ruby','C']
students = [56,69,45,56,70]
```

```python
plt.xlabel('Course Name')
plt.ylabel('No. of Students Enrolled')
plt.title('Student Enrollment', fontsize = 15)
plt.bar(courses, students, color = 'black', edgecolor = 'r',
       linewidth = 3, width = 0.5)
plt.grid()
```

# Scatterplot

```python
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.title('Line Plot')
plt.scatter(x, y, marker = '>')
plt.grid()
```

# Pie Chart

```python
plt.figure(figsize=(8,8))

plt.title('Pie Chart')
plt.pie(students, labels = courses, autopct = '%.2f%%', 
        radius= 0.8)
plt.legend();
```
